# Decision Tree Classification

```python
import pandas as pd
```

```python
df=pd.read_csv("exercise.csv")
df.head()
```

```python
latest_df=df.dropna()
```

```python
latest_df
```

```python
target=latest_df['Survived']
inputes=latest_df.drop(['Survived','PassengerId','Pclass','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
```

```python
inputes
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_sex=LabelEncoder()
```

```python
inputes['Sex_n']=le_sex.fit_transform(inputes['Sex'])
```

```python
inputs_n=inputes.drop(['Sex'],axis='columns')
```

```python
inputs_n
```

```python
target
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_n,target)
```

```python
model.score(inputs_n,target)
```
